```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import os
```

```python
data = dict()
Cp_dict = dict()
```

```python
output_dir = "../outputs_curved/output-Re{20}_u{0.15}_N0{100}_nu{0.75}_C2/"

scheme = output_dir[-3:-1]
parts = []
for left in output_dir.split('{'):
    for right in left.split('}'):
        parts.append(right)
parts[5]
```

```python
x = np.loadtxt(os.path.join(output_dir, "x.txt"))
y = np.loadtxt(os.path.join(output_dir, "y.txt"))
u = np.loadtxt(os.path.join(output_dir, "u.txt"))
v = np.loadtxt(os.path.join(output_dir, "v.txt"))
rho = np.loadtxt(os.path.join(output_dir, "rho.txt"))
errors = np.loadtxt(os.path.join(output_dir, "errors.txt"))
nx = len(x)
ny = len(y)

H = 0.41
L = 2.2
r = 0.05
x0 = 0.2
y0 = 0.2

normalize_lengths = False

N0 = int(parts[5])

dx = 2*r/N0
i0 = int(x0/dx)
j0 = int(y0/dx)

x.shape, y.shape, u.shape, v.shape, rho.shape, errors.shape
```

```python
r, dx, i0, j0
```

```python
plt.figure(figsize=(10, 7))
sns.lineplot(range(1, len(errors)+1), errors)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("iterations")
plt.ylabel("error")
plt.title("Convergence (1e-15)")

plt.show()
```

```python
P = (rho* 1/np.mean(rho))*((0.2/0.15)**2)/3

P[i0-int(r/dx), j0] - P[i0+int(r/dx), j0]
```

```python
f = np.loadtxt("./F.txt").astype(int)
angles = np.arctan2(f[:, 1]-i0, f[:, 0]-j0)*180/np.pi
surface_rhos = np.array([rho[f[i, 0]][f[i, 1]] for i in range(len(f))])
Cp_dict[scheme] = (surface_rhos-np.mean(rho[0]))/np.mean(rho[0]) * 2/(3*np.mean(u[0])*np.mean(u[0]))
```

```python
plt.figure(figsize=(16, 6))

plt.subplot(121)
plt.plot(angles[angles>=0], Cp_dict['C1'][angles>=0], color='k')
# plt.plot(angles[angles>=0], Cp_dict['C2'][angles>=0], color='b')
plt.gca().invert_xaxis()
plt.xlabel("theta (deg.)")
plt.ylabel("Cp")
plt.legend(["C1"])
# plt.legend(["C1", "C2"])
plt.title("Re = 20, N0 = 100")

plt.subplot(122)
plt.plot(angles[angles>=0], Cp_dict['bb'][angles>=0], color='g')
plt.plot(angles[angles>=0], Cp_dict['ns'][angles>=0], color='r')
plt.gca().invert_xaxis()
plt.xlabel("theta (deg.)")
plt.ylabel("Cp")
plt.legend(["BB", "NS"])
plt.title("Re = 20, N0 = 100")

plt.show()
```

```python
plt.figure(figsize=(16, 6))

plt.subplot(121)
plt.plot(angles[angles>=0], Cp_dict['C1'][angles>=0], color='k')
plt.plot(angles[angles>=0], Cp_dict['C2'][angles>=0], color='b')
plt.gca().invert_xaxis()
plt.xlabel("theta (deg.)")
plt.ylabel("Cp")
plt.legend(["C1", "C2"])
plt.title("Re = 5")

plt.subplot(122)
plt.plot(angles[angles>=0], Cp_dict['bb'][angles>=0], color='g')
plt.plot(angles[angles>=0], Cp_dict['ns'][angles>=0], color='r')
plt.gca().invert_xaxis()
plt.xlabel("theta (deg.)")
plt.ylabel("Cp")
plt.legend(["BB", "NS"])
plt.title("Re = 5")

plt.show()
```

```python
xx = 0.256
y1 = 0.13
y2 = 0.27

yy = 0.256
x1 = 0.13
x2 = 0.27

data[scheme+"v"] = P[x==xx, (y>=y1) & (y<=y2)]
data[scheme+"h"] = P[(x>=x1) & (x<=x2), y==yy]
```

```python
if normalize_lengths:
    x = x/L
    y = y/L
    r = r/L
    dx = dx/L
```

```python
cutoff = 0.22

h = 12
ar = L/H

f = plt.figure(figsize=(h*ar*cutoff, h))
c = plt.streamplot(x, y, u.T, v.T, density=15, color=np.sqrt(u.T*u.T + v.T*v.T), cmap='winter')
if normalize_lengths:
    plt.xlabel("x/L")
    plt.ylabel("y/L")
else:
    plt.xlabel("x")
    plt.ylabel("y")

plt.title("Streamlines")

plt.gca().add_patch(plt.Circle((i0*dx, j0*dx), r, color='k'))
plt.xlim([min(x), cutoff*max(x)])

cbaxes = f.add_axes([0.11, 0.04, 0.8, 0.03]) 
cb = plt.colorbar(cax=cbaxes, orientation="horizontal")

plt.show()
```

```python
cutoff = 0.22

X, Y = np.meshgrid(x, y)

f = plt.figure(figsize=(h*ar*cutoff, h))
c = plt.contour(X.T, Y.T, P, np.arange(np.min(P), np.max(P), step=0.004), cmap="winter")
if normalize_lengths:
    plt.xlabel("x/L")
    plt.ylabel("y/L")
else:
    plt.xlabel("x")
    plt.ylabel("y")
    
plt.title("Density")

plt.gca().add_patch(plt.Circle((i0*dx, j0*dx), r, color='k'))
plt.xlim([min(x), cutoff*max(x)])

# solid_points = np.loadtxt("./solid_points.txt").astype(int)
# plt.scatter(solid_points[:, 0]*dx, solid_points[:, 1]*dx)

plt.vlines(xx, y1, y2)
plt.hlines(yy, x1, x2)

cbaxes = f.add_axes([0.11, 0.04, 0.8, 0.03]) 
cb = plt.colorbar(cax=cbaxes, orientation="horizontal")

plt.show()
```
