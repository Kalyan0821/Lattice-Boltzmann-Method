```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
N0 = [50, 100, 200]

c1_P = [0.136, 0.136, 0.136]
bb_P = [0.138, 0.138, 0.137]
ns_P = [0.129, 0.129, 0.129]
P = 0.117

c1_L = [0.0896, 0.0931, 0.0948]
bb_L = [0.0896, 0.0913, 0.0913]
ns_L = [0.0827, 0.0844, 0.0827]
L = 0.0847
```

```python
plt.figure(figsize=(8, 5))
plt.plot(N0, c1_P)
plt.plot(N0, bb_P)
plt.plot(N0, ns_P)
plt.hlines(P, xmin=N0[0], xmax=N0[2], linestyle="-.")
plt.legend(["C1", "BB", "NS", "Benchmark"], edgecolor='k')
plt.xlabel("N0")
plt.ylabel("del.P (Pa)")

plt.xticks([50, 100, 150, 200])
plt.ylim([0, 0.145])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(N0, c1_L)
plt.plot(N0, bb_L)
plt.plot(N0, ns_L)
plt.hlines(L, xmin=N0[0], xmax=N0[2], linestyle="-.")
plt.legend(["C1", "BB", "NS", "Benchmark"], edgecolor='k')
plt.xlabel("N0")
plt.ylabel("del.L (m)")

plt.xticks([50, 100, 150, 200])
plt.ylim([0, 0.1])
plt.show()
```
